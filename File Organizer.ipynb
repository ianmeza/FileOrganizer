```python
#ORIGINAL POR: Mahesh Sawant
#ADAPTACIÓN: Ian Meza
#10/Mayo/2022
#


import os,shutil

s=os.chdir(f"C:/Users/62357/Downloads/")
current = os.getcwd()

files=os.listdir(current)

images=[".jpeg",".png",".jpg",".gif", ".tiff"] #extensions for images
text=[".doc",".txt",".pdf",".xlsx",".docx",".xls",".rtf", ".pptx", ".ppt", ".xls", ".csv"] #extensions for text files
videos=[".mp4",".mkv", ".flv", ".avi", ".wmv"] #extensions for videos
sounds=[".mp3",".wav",".m4a"] #extensions for sounds
applications=[".exe",".lnk", ".msi", ".msu"] #extensions for applications
codes = [".c",".py",".java",".cpp",".js",".html",".css",".php"] #extensions for codes
rdp = [".rdp"] #extensions for remote desktop
certs = [".cer", ".pem", ".pfx"] #extensions for certificates
zipped = [".rar", ".7z", ".zip"] #extensions for compressed files

print("Organizando archivos...")

for file in files:
    dest = f"C:/Users/62357/Downloads/"
    for ex in images:
        if file.endswith(ex):
            dest="Imagenes"
            shutil.move(file,dest)
            break

    for ex in text:
        if file.endswith(ex):
            dest="Documentos"
            shutil.move(file,dest)
            break

    for ex in sounds:
        if file.endswith(ex):
            dest="Sonidos"
            shutil.move(file,dest)
            break

    for ex in videos:
        if file.endswith(ex):
            dest="Videos"
            shutil.move(file,dest)
            break

    for ex in applications:
        if file.endswith(ex):
            dest= "Aplicaciones"
            shutil.move(file,dest)
            break

    for ex in codes:
        if file.endswith(ex):
            dest= "Codigos"
            shutil.move(file,dest)
            break
            
    for ex in rdp:
        if file.endswith(ex):
            dest="RDP"
            shutil.move(file,dest)
            break

    #if dest == f"C:/Users/62357/Downloads/":
    #    shutil.move(file,"Otros")

print("Organización finalizada...")
```
